```python
attack_type_train = df_train.attack.map(lambda x: 0 if x=='normal' else 1)
attack_type_test = df_test.attack.map(lambda x: 0 if x=='normal' else 1)
```

```python
df_train['attack_type'] = attack_type_train
df_test['attack_type'] = attack_type_test
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
dos_attacks = ['apache2','back','land','mailbomb','neptune','pod','processtable','smurf','teardrop','udpstorm']
probe_attacks = ['ipsweep','mscan','nmap','portsweep','saint','satan']
r2l_attacks = ['ftp_write','guess_passwd','imap','multihop','named','phf','sendmail','snmpgetattack','snmpguess','warezclient','warezmaster','worm','xlock','xsnoop']
u2r_attacks = ['buffer_overflow','httptunnel','loadmodule','perl','ps','rootkit','sqlattack','xterm']
```

```python
def attack_class (attack):
    if attack in dos_attacks:
        return 1
    elif attack in probe_attacks:
        return 2
    elif attack in r2l_attacks:
        return 3
    elif attack in u2r_attacks:
        return 4
    else:
        return 0
```

```python
attack_multi_type_train = df_train.attack.apply(attack_class)
attack_multi_type_test = df_test.attack.apply(attack_class)
```

```python
df_train['attack_multi_type'] = attack_multi_type_train
df_test['attack_multi_type'] = attack_multi_type_test
```
